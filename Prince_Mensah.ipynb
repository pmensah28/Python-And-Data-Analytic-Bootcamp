# HANDELING NULL VALUES

The dataset may contains rows and columns for whichc some values are missing, we can't leave such values as it is;
In such cases we have two options.

1. Either we drop the entire rows and columns
2. Fill the missing values with some appropriate values, let's say mean of all the the specific columns.

```python
df.isnull().sum()
```

**Separating out the columns which have more than 35% of the values missing in the dataset.**

```python
# df.isnull().sum() returns a pandas series with column names as the label index 
# and total count of null values in the column as it's value.
# And we are storing only those columns which have values more than 35% of the data missing.

x = df.isnull().sum()
drop_col = x[x>(35/100 * df.shape[0])]
drop_col
```

**NOTE:**There is no specific number after which you should drop the columns. It's just that we decided that on our own according to what wee want

```python
drop_col.index
```

```python
df.drop(drop_col.index, axis=1, inplace=True)
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

Because **Embarked** contains string values, we see the details of that column separately from others as strings does not have mean and all.

```python
df['Embarked'].describe()
```

For Embarked attributes, we fill the NULL values with the most frequent value in the column.

```python
df['Embarked'].fillna('S', inplace=True)
```

```python
df.isnull().sum() # NOW ALL THE NULL VALUES HAVE BEEN FILLED
```

```python
df.corr()
```

**sibsp** = Number of Siblings/Spouses Aboard
**parch** = Number of Children/Children 

So we can make a new column family_size by combining these two columns

```python
df['FamilySize'] = df['SibSp']+df['Parch']
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df.corr()
```

**FamilySize in the sheep does not have much correlance with survival rate.**

Let's check if the person was alone or not could affect survival rate.

```python
df['Alone'] = [0 if df['FamilySize'][i]>0 else 1 for i in df.index]
df.head()
```

```python
df.groupby(['Alone'])['Survived'].mean()
```

If the person is alone he or she has less chance of surving.

The reason might be the person who was travelling with his family might belong to rich class and might be patronized over there.

```python
df[['Alone', 'Fare']].corr()
```

So we can check that if the person was not alone there is high chance that the ticket price will be high.

```python
df['Sex'] = [0 if df['Sex'][i]=='male' else 1 for i in df.index]  # 1 for female and 0 for male
df.groupby(['Sex'])['Survived'].mean()
```

It shows female passengers had high chance of surviving than male ones.

It shows women were priotirized over men

```python
df.groupby(['Embarked'])['Survived'].mean()
```
